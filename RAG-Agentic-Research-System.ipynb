```python
import urllib.request
from pathlib import Path

for url, paper in zip(urls, papers):
    if not Path(paper).exists():
        print(f"Downloading {paper}...")
        urllib.request.urlretrieve(url, paper)
    else:
        print(f"{paper} already exists")
```
